### 可视化不同算子下，图像A和B的psnr

```python

import matplotlib.pyplot as plt
import pandas as pd
```

```python
siren_fit_path = 'SIREN_fit_exp_task10.csv'
siren_add_path = 'SIREN_add_exp_task10.csv'
siren_sub_path = 'SIREN_sub_exp_task10.csv'
siren_mul_path = 'SIREN_mul_exp_task10.csv'

df_siren_fit = pd.read_csv(siren_fit_path)
df_siren_add = pd.read_csv(siren_add_path)
df_siren_sub = pd.read_csv(siren_sub_path)
df_siren_mul = pd.read_csv(siren_mul_path)

df = pd.DataFrame()
df['task'] = df_siren_fit['task'].copy()
df['fit_A'] = df_siren_fit['psnrA'].copy()
df['add_A'] = df_siren_add['psnrA'].copy()
df['sub_A'] = df_siren_sub['psnrA'].copy()
df['mul_A'] = df_siren_mul['psnrA'].copy()

df['fit_B'] = df_siren_fit['psnrB'].copy()
df['add_B'] = df_siren_add['psnrB'].copy()
df['sub_B'] = df_siren_sub['psnrB'].copy()
df['mul_B'] = df_siren_mul['psnrB'].copy()


df.head()

# df.plot(x='epoch', y=['train/psnr1', 'train/psnr2'], kind='line', grid=True)
```

```python
df.sum() / df.shape[0]
```

```python
df.plot(x='task', y=['fit_A', 'add_A', 'sub_A', 'mul_A'], kind='line', grid=True)
```

```python
df.plot(x='task', y=['fit_B', 'add_B', 'sub_B', 'mul_B'], kind='line', grid=True)
```
