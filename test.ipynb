### 可视化不同算子下，图像A和B的psnr

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# metrics_path = '/data/liuzhen/meta_learning_data/DataSets/operator_tasks_celea_face_100_178/task_000000/logs/SIREN_fit_exp_1/version_0/metrics.csv'
# fig, ax = plt.subplots(3,3)
TITLE='pluscosE'
TASK_NAME='exp_lr0.001_MLP+PE_'+TITLE
psnr1=[]
psnr2=[]
for i in range(9):
    pos='33'+str(i+1)
    plt.subplot(pos)
    metrics_path = '/home/hashnerf/nerfpl/DataSet/data/logs/'+TASK_NAME+'/version_'+str(i)+'/metrics.csv'

    df = pd.read_csv(metrics_path)

    print(metrics_path)
    psnr1.append(df['train/psnr1'].max())
    psnr2.append(df['train/psnr2'].max())
    plt.plot(df['epoch'], df['train/psnr1'],label = 'psnr1')
    plt.plot(df['epoch'], df['train/psnr2'],label = 'psnr2')
    plt.grid(linestyle="--", alpha=0.8, zorder=0)
    plt.tick_params(labelsize=6)

    # df.plot(x='epoch', y=['train/psnr1', 'train/psnr2'], kind='line', grid=True)
print('psnr1: ', np.mean(psnr1))
# print('psnr2: ', psnr2)
print('psnr2: ', np.mean(psnr2))
plt.legend() 
plt.suptitle(TITLE)
plt.tick_params(labelsize=6)
plt.savefig("paper_figs/"+TASK_NAME+".pdf",bbox_inches = 'tight',format='pdf')
```
